<p class="m-0 p-0">The first thing I'd do is create variables to hold the
<ul class="">
    <li>"apple" string  --  held by <code class="underlined">a</code></li>
    <li>The returned <code>input()</code> string  --  held by <code class="underlined">one</code></li>
</ul>
</p>

```python
a = "apple"
one = None
```

<p>I'll initially set the variable <code class="underlined">one</code> to value of <code>None</code>, so that I know to change it later (because <code>None</code> is usually useless for calculations and logical operations).</p>
<p>Then I'll setup a <code>while</code> loop, that will run infinitely (by using <code>while True:</code>) until I instruct it to <code>break</code> out of it.</p>

```python
while True:
    one = input("Please enter your password: ")
```

<p>Time to compare with <code>if</code> statements.<br>
Setup an <code>if</code> that checks whether <code class="underlined">a</code> and <code class="underlined">one</code> have the same value (are the same), then <code>break</code> out of the loop if so.</p>
<p>If it's not the same, meaning it isn't the correct password, allow them to suffer perpetual Hell as they try and fail to guess the password.<br>
The best part about loops is that if you want said loop to repeat, you don't have to do anything, after all, it runs infinitely until we manually <code>break</code> it!</p>

```python
    if one == a:
        print("Correct")
        break
    else:
        print("Uh oh, stinky \n")
    # The while loop will loop-over again
```

<p>And after that <code>if else</code>, you're essentially finished with your program!<br>
You can give your user a Cookie for correctly guessing the password.</p>

```python
# Outside of the while loop
print("\n You win... a Cookie!")
```

<h5>Bringing it all together</h5>
<p>Feel free to copy'n'paste this into your Python IDLE to see if it actually works, I assure you, it does work <span class="text-warning">:)</span></p>

```python
a = "apple"
one = None
while True:
	one = input("Please enter your password: ")
	if one == a:
		print("Correct!")
		break
	else:
		print("Incorrect, try again\n")
	# The while loop will loop-over again
# Outside of while loop
print("\n You win... a Cookie!")
```
